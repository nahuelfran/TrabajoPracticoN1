Decidi eliminar los valores atipicos del data set ya que estos valores atípicos pueden ser datos inusuales o errores de medición que pueden distorsionar nuestros análisis La eliminación de estos valores nos permite trabajar con un conjunto de datos más consistente y confiable.

```python
filas,columnas=df.shape
print("\n El número de muestras en el dataframe es ",filas*columnas,"\n\n")


print(f"\n El dataframe tiene {filas} filas y {columnas} columnas\n\n")
print(df.info())

# - LIMPIEZA - MANEJO DE VALORES NULOS:
#verificar la cantidad de valores nulos en cada columna
n_nulos = df.isnull().sum()
print(f'\n\nel número de nulos es:\n\n {n_nulos}\n')

# data set limpio
df_clean = df.dropna()
n_nulos = df_clean.isnull().sum()
print(f'\n\nel número de nulos el data frame limpio es:\n\n {n_nulos}\n')

# - LIMPIEZA - MANEJO DE VALORES ATIPICOS:
#lista de columnas con los datos a determinar valores atipicos
columnas_interes = columnas_interes = ['world_rank', 'national_rank', 'quality_of_education', 'alumni_employment', 'quality_of_faculty', 'publications', 'influence', 'citations', 'broad_impact', 'patents', 'score']


for columna in columnas_interes:
    # Visualización de datos antes de la limpieza
    plt.figure(figsize=(8, 6))
    sns.histplot(df_clean[columna], kde=True)
    plt.title(f"Histograma de la columna {columna} (Antes de la limpieza)")
    plt.xlabel(columna)
    plt.ylabel("Frecuencia")
    plt.show()

    # Estadísticas descriptivas para la columna actual
    estadisticas = df_clean[columna].describe()
    media = estadisticas['mean']
    std = estadisticas['std']

    # Identificación de valores atípicos utilizando un umbral basado en la desviación estándar (ajusta el umbral según sea necesario)
    umbral = 2 * std
    valores_atipicos = df_clean[abs(df_clean[columna] - media) > umbral]

    # Visualización de valores atípicos antes de la eliminación
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df_clean[columna])
    plt.title(f"Diagrama de Caja para Identificar Valores Atípicos en la columna {columna} (Antes de la limpieza)")
    plt.xlabel(columna)
    plt.show()

    # Eliminar valores atípicos en df_clean
    df_clean = df_clean[abs(df_clean[columna] - media) <= umbral]

# Muestra el DataFrame limpio después de la eliminación de valores atípicos
print("DataFrame limpio después de eliminar valores atípicos:")
print(df_clean.head())
```

##***Ejercicio 7***
Sobre el Dataset Elegido  explique si se puede reducir las dimensiones  y que representa esas  nuevas variables.

Después de realizar un análisis de Componentes Principales (PCA) en nuestro conjunto de datos, he observado que es posible reducir significativamente la dimensionalidad. Originalmente, teníamos un conjunto de datos con múltiples variables que describen las características de las universidades. Sin embargo, al aplicar PCA, hemos generado nuevas variables llamadas 'Componente Principal 1', 'Componente Principal 2', que son combinaciones lineales de las variables originales. Estas nuevas variables (componentes principales) retienen la mayoría de la varianza en los datos originales.

La reducción de dimensiones nos permite simplificar la representación de nuestros datos mientras conservamos la información esencial. En particular, hemos identificado que los primeras siete componentes principales explican aproximadamente 95% de la varianza total en los datos. Esto sugiere que podemos trabajar con un conjunto de datos de dimensionalidad reducida mientras mantenemos una gran parte de la información original.

Estas nuevas variables, es decir, los componentes principales, representan combinaciones ponderadas de las características originales y pueden interpretarse como dimensiones latentes que capturan patrones y estructuras subyacentes en nuestros datos. Esto simplifica el análisis y puede ayudarnos a identificar relaciones y tendencias ocultas entre las universidades que no eran evidentes en las variables originales.

```python
## Creamos un Dataframe para valores estandarizados
df_scaled = df_clean
# Estandarizar los datos
scaler = StandardScaler()
df_scaled[columnas_interes] = scaler.fit_transform(df_clean[columnas_interes])

# Aplicar PCA sin especificar n_componentes inicialmente
pca = PCA(random_state=42)
pca_df = pca.fit_transform(df_scaled[columnas_interes])

# Obtener la varianza explicada acumulativa
explained_variance_ratio_cumulative = np.cumsum(pca.explained_variance_ratio_)

# Graficar la varianza explicada acumulativa
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio_cumulative) + 1), explained_variance_ratio_cumulative, marker='o')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulativa')
plt.title('Varianza Explicada Acumulativa por Componentes Principales')
plt.grid(True)

# Determinar el número de componentes que explican una cantidad deseada de varianza
varianza_deseada = 0.95
n_componentes_deseados = np.argmax(explained_variance_ratio_cumulative >= varianza_deseada) + 1

plt.axvline(x=n_componentes_deseados, color='red', linestyle='--', label=f'95% Varianza (n={n_componentes_deseados})')
plt.legend()
plt.show()

# Usar el número de componentes deseados en PCA
pca = PCA(n_components=n_componentes_deseados, random_state=42)
pca_df = pca.fit_transform(df_scaled[columnas_interes])

# Visualizar los datos en el espacio de componentes principales
plt.figure(figsize=(10, 6))
plt.scatter(pca_df[:, 0], pca_df[:, 1], c=df_scaled['world_rank'].astype('category').cat.codes, cmap='viridis')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('PCA: Espacio de Componentes Principales')
plt.colorbar(label='world_rank')
plt.show()
```
